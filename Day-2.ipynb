```python
f=open('marks.txt')
f1=open('total marks.txt','w')
d=f.readlines()
for i in d:
    words=i.split()
    name=words[0]
    #total = int(words[1])+int(words[2])+int(words[3])+int(words[4])
    total= sum(list(map(int,words[1:])))
    print(name,"",total)
    f1.write(name+""+str(total)+"\n")
f1.close()
    
```

## Regular Expressions

```python
import re
```

```python
#re.match(pattern,string)
#re.search()
#re.findall()
```

```python
a="bbaSaiRam"
re.match('b',a)
```

```python
a="AbbaSaiRam"
re.search('b',a
         )
```

```python
a="AbbaSaiRam"
re.findall('b',a)
```

```python
# ^ matches the beginning of the line
a="AbbaSaiRam"
re.search('^b',a)
```

```python
# $ matches the ending of a line
a="AbbaSaiRam"
re.search('m$',a)
```

```python
a="abbaSAira,"
re.search('.',a)
```

```python
a=['ravi','anil','raju','ranga','uma','sai']
for i in a:
    if re.search('^r.{4,6}$',i):
        print(i)
```

```python
# \d matches the digits only
a="gaya31gaya312"
re.findall('\d+',a)
```

```python
# \D matches the non-digits only
a="@gaya3$"
re.search('\D+',a)
```

```python
# \W matches the special characters
a="gaya3$"
re.findall('\W*',a)
```

```python
# \s matches the Alphabets
a="gaya  3"
re.search('\s+',a)
```

```python
# [a-z] alphabe5ts in listed set
re.search('[0-9]+','ravi  0559')
```
